 # Task 1: Exploring and Visualizing the Iris Dataset

  ### Objective: Explore and visualize Iris flower data for business insights


### Step 1: Import Libraries

```python
import seaborn as sns
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

### Step 2: Load the Dataset

```python
 
# Load the built-in Iris dataset from seaborn
df = sns.load_dataset('iris')
```

# -------------------------------
# 1. Dataset Structure Overview
# -------------------------------

```python
print("Shape of the dataset:", df.shape)
print("\nColumn names:", df.columns)
print("\nFirst 5 rows:")
print(df.head())
```

# -------------------------------
# 2. Scatter Plot: Petal Length vs Petal Width
### Use-case: Understand bloom size across species
# -------------------------------

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df, s=100)
plt.title('Bloom Size by Petal Dimensions')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.grid(True)
plt.show()
```

# -------------------------------
# 3. Histogram: Sepal Length Distribution
### Use-case: Understand height preferences for bouquet packaging
# -------------------------------

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='sepal_length', hue='species', kde=True, palette='pastel', bins=15)
plt.title('Height Distribution by Species (Sepal Length)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Count')
plt.show()
```

# -------------------------------
# 4. Box Plot: Sepal Width by Species
### Use-case: Detect species with wider flowers for packaging concerns
# -------------------------------

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='species', y='sepal_width', palette='Set2')
plt.title('Width Comparison for Packaging (Sepal Width)')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

# -------------------------------
# 5. Pairplot: Overall Feature Relationships
### Use-case: Business summary for flower traits across species
# -------------------------------

```python
sns.pairplot(df, hue='species', palette='husl')
plt.suptitle('Pairwise Relationships Between Flower Features', y=1.02)
plt.show()
```
